```python
import pandas as pd
import numpy as np
```

```python
fileb="usbirth.csv"
filep="uspop.csv"
filed="usdeaths.csv"
filedivorce="usdivorce.csv"

birth=pd.read_csv(fileb)
pop=pd.read_csv(filep)
death=pd.read_csv(filed)
divorce=pd.read_csv(filedivorce)
```

```python
newbirth=birth.loc[0:69]
pbirth=birth.loc[70:]
newpop=pop.loc[0:69]
ppop=pop.loc[70:]
newdeath=death.loc[0:69]
pdeath=death.loc[70:]
newbirth=newbirth.drop(columns=[" Annual % Change"])
newdeath=newdeath.drop(columns=[" Annual % Change"])
newpop=newpop.drop(columns=[" Annual % Change"])
bp=pd.merge(newbirth,newpop,how='left',on="date")

bp.head()

bp["Total Number of Births"]=((bp[" Births per 1000 People"]*bp[" Population"])/1000)

bp.head()

bpd=pd.merge(bp,newdeath,how='left',on='date')

bpd.head()

bpd["Total Number of Deaths"]=((bpd[" Deaths per 1000 People"]*bpd[" Population"])/1000)
bpd.insert(1,"Country","United States")
bpd.head()
```

```python
divorce.tail()
```

```python
uscomplete=bpd.merge(divorce,how='left',on='date')
uscomplete.head()
```

```python
complete_us_df=uscomplete.to_csv("Complete US DF.csv")
us_no_divorce_stats=bpd.to_csv("DataFrame without Divorce.csv")
```
